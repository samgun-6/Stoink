```python
#imports
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
import http.client
import json as json
```

```python
# Loading the data from CSV into panadas dataframe
df = pd.read_csv (r'Toymodelstockdata\stocksdatatoymodel.csv', sep=',')
#fashion_train_df = pd.read_csv('input/fashion-mnist_train.csv',sep=',')

# printing
print (df)
```

```python
# Adding artificial labels
# Randomizing numbers to represent the increase in value in %
df['Performance'] = np.random.randint(-10, 30, size=len(df))
print(df)
print(len(df))
```

```python
# Making the toy model with tensorflow
model = tf.keras.models.Sequential()

# Layers
model.add(tf.keras.layers.Dense(units = 32, activation = 'relu', input_dim=4))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
# compiling the model
model.compile(loss = 'binary_crossentropy', optimizer = 'sgd')
```

```python
# Slicing the data
#print(df.iloc[:,-1:])
target = df.iloc[:,-1:].to_numpy()

column_features = ["totalAssets","inventory","totalShareholderEquity","retainedEarnings"]
features = df[column_features].to_numpy()

# printing
print(target)
print(features)
```

```python
# fitting the model
model.fit(features, target, epochs = 10, batch_size = 32)
```
