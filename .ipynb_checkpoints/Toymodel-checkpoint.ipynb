```python
#imports
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
import http.client
import json as json
from tensorflow import keras
```

```python
# Loading the data from CSV into panadas dataframe
df = pd.read_csv (r'Toymodelstockdata\stocksdatatoymodel.csv', sep=',')
#fashion_train_df = pd.read_csv('input/fashion-mnist_train.csv',sep=',')

# printing
print(df)
```

```python
# Adding artificial labels
# Randomizing numbers to represent the increase in value in %
df['Performance'] = np.random.randint(0, 40, size=len(df))
print(df)
print(len(df))
```

```python
# Making the toy model with tensorflow
model = tf.keras.models.Sequential()

# Layers
model.add(tf.keras.layers.Dense(units = 4, activation = 'relu', input_dim=4))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
# compiling the model
opt = keras.optimizers.Adam(learning_rate=0.005)
model.compile(loss = 'binary_crossentropy', optimizer = opt)
```

```python
# Slicing the data
#print(df.iloc[:,-1:])
target = df.iloc[:,-1:].to_numpy()

column_features = ["totalAssets","inventory","totalShareholderEquity","retainedEarnings"]
features = df[column_features].to_numpy()

# Normalizing features
#features = features/1000000000
row_sums = features.sum(axis=1)
norm_features = features / row_sums[:, np.newaxis]
# adding intercept term
#m, n = features.shape
#features = np.concatenate([np.ones((m, 1)), features], axis=1)


# printing
#print(target)
#print(features)
print(norm_features)
```

```python
# fitting the model
model.fit(norm_features, target, epochs = 100, batch_size = 32)
```
